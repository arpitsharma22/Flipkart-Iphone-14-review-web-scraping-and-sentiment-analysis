```python

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns 
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.tokenize import TweetTokenizer
from sklearn.feature_extraction.text import CountVectorizer, TfidfTransformer
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.model_selection import train_test_split
import requests
from bs4 import BeautifulSoup
import re
```

```python
reviews = []
comments = []
ratings = []
for i in range(1,180):
    url = 'https://www.flipkart.com/apple-iphone-14-midnight-128-gb/product-reviews/itm9e6293c322a84?pid=MOBGHWFHECFVMDCX&lid=LSTMOBGHWFHECFVMDCXSSCYWA&marketplace=FLIPKART&page='+str(i)
    r = requests.get(url)
                       
    soup = BeautifulSoup(r.text,'lxml')
    
    review = soup.find_all('p','_2-N8zT')
    comment = soup.find_all('div','t-ZTKy') 
    rating = soup.find_all("div","_3LWZlK _1BLPMq")
    
    reviews.append(review)
    comments.append(comment)
    ratings.append(rating)
```

```python
print(len(reviews))
print(len(comments))
print(len(ratings))
```

```python
reviews
```

```python
rev = []





for i in reviews:
    for j in i:
        t = j.text
        rev.append(t)
```

```python
comments
```

```python
com = []





for i in comments:
    for j in i:
        t = j.text.replace("READ MORE"," ")
        com.append(t)
```

```python
ratings
```

```python

rate = []


for i in ratings:
    for j in i:
        t = j.text
        rate.append(t)
```

```python
cleaned_comment_list = [re.sub(r'[^a-zA-Z ]', '', comments) for comments in com]
cleaned_comment_list
```

```python
print(len(rate))
print(len(cleaned_comment_list))
print(len(rev))
```

```python
if len(cleaned_comment_list) != len(rev):
    raise ValueError("Length of comments and reviews do not match.")
elif len(cleaned_comment_list) != len(rate):
    rate.extend([5] * (len(cleaned_comment_list) - len(rate)))
```

```python
print(len(rate))
```

```python
data=pd.DataFrame({"Reviews": rev,"Comment":cleaned_comment_list,'Ratings':rate})
data.shape
```
